#### 1. Initialization and Tensor Hierarchies(Rank)
- TensorFlow use `tf.constant` for create tensor constant (its value does not change after creation).

    - `scalar = tf.constant(7)` # rank-0
        - This is a single number. It has no dimension (like a point). In mathematics, this is called a scalar.
        - Shape: () (blank).

    - `vector = tf.constant([10, 20, 30])` # rank-1
        - A list of numbers. It has one dimension (horizontal or vertical).
        - Shape: (3,).

    - `matrix = tf.constant([[1., 2.], [3., 4.]])` # rank-2
        - A table of numbers has rows and columns. This is a 2D array.
        - Shape: (2, 2).


    - `tensor = tf.constant([[[1],[2]], [[3],[4]]])` # rank-3 (2x2x1)
        - imagine this as a cube of data or a set of matrices stacked on top of each other.
        - Shape: (2, 2, 1).

```python
# 1) Scalars, vectors, matrices, tensors
scalar = tf.constant(7) # rank-0
vector = tf.constant([10, 20, 30]) # rank-1
matrix = tf.constant([[1., 2.], [3., 4.]]) # rank-2
tensor = tf.constant([[[1],[2]], [[3],[4]]]) # rank-3 (2x2x1)
print(scalar.shape, vector.shape, matrix.shape, tensor.shape)
```

#### 2. Data Types and Type Casting (ép)

- In Deep Learning, data type control is extremely important for optimizing memory usage and computation speed.
    - `x = tf.constant([1.7, 7.4])` # float32 default
    - `y = tf.constant([7, 10])` # int32 default
    - By default, if I enter a floating-point number, TF selects `float32`. If you enter an integer, TF selects `int32`.

- `x16 = tf.cast(x, tf.float16) # reduced precision floats`

    - **tf.cast:** Used for data type conversion. Converting to `float16` reduces memory usage by half (commonly used in "Mixed Precision" techniques to train models faster on the GPU).

- `y32 = tf.cast(y, tf.float32)` # cast ints → floats

    - Convert integers to floating-point numbers. Note: In TensorFlow, you **cannot** perform math between a tensor `int32` and a `float32`. I must `cast` them to the same type.

TensorFlow often uses float32 for deep learning.

Float16 is lighter and faster (especially on GPUs) but less accurate.

int32 is commonly used for indexes/labels; to calculate mean/gradients, it's usually necessary to convert to float.

```python
# 2) Dtypes and casting
x = tf.constant([1.7, 7.4]) # float32 default
y = tf.constant([7, 10]) # int32 default
x16 = tf.cast(x, tf.float16) # reduced precision float
# y16 = tf.cast(y, dtype=tf.float16)
y32 = tf.cast(y, tf.float32) # cast ints → floats

print(x.shape,y.shape)
print(x16.dtype)

#tf.cast: Used to convert data types.
```

#### 3. Chỉ mục (Indexing) và Thay đổi hình dạng (Shaping)

- `rank2 = tf.constant([[10, 7], [3, 4]])`
- `last_col = rank2[:, -1] # column slice`

- **Syntax `[row, column]`:**
    - The `:` means get all rows
    - `1` get final column
    - **result:** turn back tensor `[7, 4]`.


- `expanded = tf.expand_dims(rank2, -1) # add channel-like axis`

    - **tf.expand_dims:** Adds a new dimension to the tensor without changing the data inside.
    - **-1:** add in last position
    - Why need to add?? Deep Learning models (like CNNs for images) often require data in the format `(Batch, Height, Width, Channels)`. If you have a black and white image `(28, 28)`, you must use this command to transform it into `(28, 28, 1)` so that the model can recognize it.



```python
# 3) Indexing and shaping
rank2 = tf.constant([[10, 7], [3, 4]])
last_col = rank2[:, -1] # column slice, #Get all rows in the last column
expanded = tf.expand_dims(rank2, -1) # add channel-like axis, Adds a new dimension to a tensor (often used to add a "channel" dimension to an image).
squeezed = tf.squeeze(expanded, axis=-1) #Remove dimensions of size 1 to simplify the tensor.

print(rank2.shape)
print(last_col)
print(f'Expanded:{expanded}')
print(f'Sq:{squeezed}')
```

#### 4. Broadcasting 
- is an incredibly clever mechanism that allows you to perform calculations between Tensors of different sizes without needing to write loops

```python
# 4) Broadcasting & basic ops
A = tf.constant([[10., 7.], [3., 4.]]) 
print("A+10:\n", A + 10)
print("A*2:\n", A * 2)
```

#### 5.Matrix mul
- (A x B) @ (B x C) 
- (3x2) @ (3x2)
- 2  ≠  3

```python
# 5) Matrix multiplication rules (inner dims match)
X = tf.constant([[1, 2], [3, 4], [5, 6]], dtype=tf.float32) # (3x2)
Y = tf.constant([[7, 8], [9, 10], [11, 12]], dtype=tf.float32) # (3x2)
XtY = tf.matmul(X, tf.transpose(Y)) # (3x2) @ (2x3) -> (3x3)

print(XtY)
```

#### 6. Aggreation
- In TensorFlow, aggregate functions (sum, average, etc.) usually begin with the keyword `reduce`.

```python
# 6) Aggregations
E = tf.constant(np.random.randint(0, 100, size=10))
print(E)
print("min:", tf.reduce_min(E).numpy(),
"max:", tf.reduce_max(E).numpy(),
"mean:", tf.reduce_mean(tf.cast(E, tf.float32)).numpy(),
"sum:", tf.reduce_sum(E).numpy())
```
