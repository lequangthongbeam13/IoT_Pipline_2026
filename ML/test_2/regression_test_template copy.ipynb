```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

df = pd.read_csv('004t_20260313-135854.csv', encoding="latin1")
df.head()

df_long = df.melt(id_vars=['Area','Age'], var_name='Year', value_name='Population')
df_long['Year'] = df_long['Year'].str.extract(r'(\d+)').astype(int)
df_long.head()

age_group = 'age 0-6'
df_age = df_long[df_long['Age'] == age_group]
X = df_age[['Year']]
y = df_age['Population']
```

## 5. Train-Test Split (YOUR WORK)

```python
# Write your code here
import tensorflow as tf

X_train, y_train = X[:15], y[:15]
X_test, y_test = X[15:], y[15:]
# 1) CREATE
model = tf.keras.Sequential([
tf.keras.layers.Input(shape=(1,)),
tf.keras.layers.Dense(1) # linear output for regression
])
# 2) COMPILE
model.compile(
loss="mae",
optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
metrics=["mae"]
)
# 3) FIT
# Let the model view the data set Train repeated 200 times (epochs=200).
history = model.fit(X_train, y_train, epochs=400, verbose=0)
# Evaluate
loss, mae = model.evaluate(X_test, y_test, verbose=0)
print("Test MAE:", mae)
# Predict and visualize
y_pred = model.predict(X_test).squeeze()
plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train, c='b', label='Train')
plt.scatter(X_test, y_test, c='g', label='Test')
plt.scatter(X_test, y_pred, c='r', label='Pred')
plt.legend(); plt.title("Regression: y = x + 10 learned")
plt.show()
```
