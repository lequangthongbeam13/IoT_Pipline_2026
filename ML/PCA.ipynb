Principal Component Analysis (PCA) is a dimensionality reduction technique.

It is used to:

Reduce the number of features in a dataset

Preserve the most important information (variance)

#### The Wine dataset contains chemical analysis of wines.

Dataset characteristics: 178 samples, 13 chemical features, 3 wine classes

Example features: Alcohol, Malic acid, Ash, Magnesium, Phenols

and this use PCA to reduce 13 dimensions to 2 dimensions for visualization.

```python
import pandas as pd

df = pd.read_csv("wine.csv")
df.head()
```

```python
df.columns
```

```python
X = df.drop("Wine", axis=1)
y = df["Wine"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_pca[:,0], X_pca[:,1], c=y)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Wine Dataset")
plt.show()
```

```python
print(pca.explained_variance_ratio_)
```

The PCA projection shows the dataset in two dimensions:

PC1 explains 36.2% of the variance

PC2 explains 19.2% of the variance

Together they represent about 55% of the original information.

The plot clearly shows three clusters, representing the three wine classes.

PCA for Image Compression

Images contain very high-dimensional data.
PCA can reduce the number of dimensions while keeping most of the important information.

Benefits:

smaller storage size

faster processing

Example:

1000 × 1000 image
= 1,000,000 features

Using PCA:

1,000,000 → 100 components

The image still looks similar but requires much less data.
